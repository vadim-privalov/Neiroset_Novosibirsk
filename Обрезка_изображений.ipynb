```python
# load the image and show it
# загрузим изображение и покажем его
# image = cv2.imread("lion.png")
image = cv2.imread("lion.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Original")
```

```python
# cropping an image is accomplished using simple NumPy array slices --
# let's crop the face from the image
# обрезка изображения выполняется с помощью 
# простых срезов массива NumPy 
# давайте вырежем лицо с изображения
face = image[81:341, 187:397]
plt.imshow(face)
plt.title("Face")
```

```python
# ...and now let's crop the entire body
# а теперь обрежем тело
body = image[30:508, 105:724]
plt.imshow(body)
plt.title("Body")
```
