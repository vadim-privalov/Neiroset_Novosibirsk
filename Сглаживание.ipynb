```python
# import the necessary packages
# импортируем необходимые модули
import cv2
from google.colab.patches import cv2_imshow
```

```python
# construct the argument parser and parse the arguments
# создадим парсер и укажем аргументы
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--image", required=True, help="Path to the image")
# args = vars(ap.parse_args())
# пока мы используем Jupyter Notebooks мы можем заменить наш парсер
# явными аргументами и значениями
args = {
	"image": "/content/Lesson_16.Smoothing/elephants.png",

}
```

```python
# load the image, display it, and construct the list of bilateral
# filtering parameters that we are going to explore
# загружаем изображение, отображаем его и 
# составляем список параметров двусторонней фильтрации,
# которые мы собираемся исследовать
image = cv2.imread(args["image"])
print('Original')
cv2_imshow(image)
params = [(11, 21, 7), (11, 41, 21), (11, 61, 39)]
```

```python
# loop over the diameter, sigma color, and sigma space
# перебираем диаметр, сигму цвета и сигму пространства
for (diameter, sigmaColor, sigmaSpace) in params:
	# apply bilateral filtering and display the image
    # применяем двустороннюю фильтрацию и отображаем изображение
	blurred = cv2.bilateralFilter(image, diameter, sigmaColor, sigmaSpace)
	title = "Blurred d={}, sc={}, ss={}".format(diameter, sigmaColor, sigmaSpace)
	cv2_imshow(blurred)
	print(title)
```
