```python
# load the image and show it
# загрузим изображение и покажем его
image = cv2.imread("elephants.png")
cv2_imshow(image)
print("Original")
```

```python
# Masking allows us to focus only on parts of an image that interest us.
# A mask is the same size as our image, but has only two pixel values,
# 0 and 255. Pixels with a value of 0 are ignored in the orignal image,
# and mask pixels with a value of 255 are allowed to be kept. 
# Маскирование позволяет нам сосредоточиться только 
# на интересующих нас частях изображения. 
# Маска имеет тот же размер, что и наше изображение, 
# но имеет только два значения пикселей, 0 и 255. 
# Пиксели со значением 0 игнорируются в исходном изображении,
# а пиксели маски со значением 255 могут быть сохранены. 
# Например, давайте создадим прямоугольную маску, 
# отображающую только одного слона на изображении.
mask = np.zeros(image.shape[:2], dtype="uint8")
cv2.rectangle(mask, (266, 116), (788, 494), 255, -1)
cv2_imshow(mask)
print("Mask")
```

```python
# применяя маску обратите внимание, 
# как обрезан только один слон на изображении
masked = cv2.bitwise_and(image, image, mask=mask)
cv2_imshow(masked)
print("Mask Applied to Image")
```

```python
# Now, let's make a circular mask with a radius of 100 pixels and apply the
# mask again
# Теперь давайте сделаем круговую маску 
# радиусом 100 пикселей и снова применим маску
mask = np.zeros(image.shape[:2], dtype="uint8")
cv2.circle(mask, (412, 214), 167, 255, -1)
masked = cv2.bitwise_and(image, image, mask=mask)
cv2_imshow(mask)
print("Mask")
```
