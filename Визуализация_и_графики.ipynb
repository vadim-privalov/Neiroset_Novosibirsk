### Настройка графика: цвета и стили линий
Цвет линии можно указать задав параметр `color`, причем цвет можно указать различными способами:

````
plt.plot(x, np.sin(x  - 0), color='blue') # Задать название
plt.plot(x, np.sin(x  - 1), color='g')  # краткий код цвета (rgbcymk)
plt.plot(x, np.sin(x  - 2), color='0.75') # шкала оттенков серого (от 0 до 1)
plt.plot(x, np.sin(x  - 3), color='#FFDD44') # 16-ричнный код RRGGBB (от 00 до FF)
plt.plot(x, np.sin(x  - 4), color=(1.0, 0.5, 0.3)) # кортеж RGB (значение от 0 до 1)
plt.plot(x, np.sin(x  - 5), color='chartreuse') # любой цвет HTML
plt.show()
````

Стиль линий можно указать с помощью параметра `linestyle`

````
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
# те же стили в сокращенном варианте:
plt.plot(x, x + 4, linestyle='-')
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 6, linestyle='-.')
plt.plot(x, x + 7, linestyle=':')
````

Можно задать цвет, стиль и толщину линии в третьем позиционном параметре:

````
plt.plot(x, x + 0, '-g4') # solid, green, 4
plt.plot(x, x + 1, '--c4') # dashed, cyan, 1
plt.plot(x, x + 2, '-.k') # dashdot, black, 1
plt.plot(x, x + 3, ':r') # dotted, red, 1
````

### Диаграммы рассеяния
что понимается под диаграммой рассеяния проще понять из примера:

````
x = np.linspace(0, 10, 31)
y = np.sin(x)
plt.plot(x, y, 'o', color='black')
````

В диаграмме рассеяния точки не соединяются линиями, а отображаются отдельными символами - *маркерами*.

Выведем список возможных маркеров:

````
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd', 'p']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker)) # rng.rand(5) создает список из 5 случайных чисел от 0 до 1
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
````

Маркеры можно использовать совместно с непрерывными линиями:

````
plt.plot(x, y, '-ok') # линия (-), маркер круг (o), черный цвет (k)
````

С помощью дополнительных именованых параметров можно задать свойства маркеров и линий:

````
plt.plot(x, y, '-p', color='gray', markersize=15, linewidth=4, markerfacecolor='yellow', markeredgecolor='green', markeredgewidth=2)
````

Для построения диаграмм рассеяния есть специальный метод `plt.scatter()`, обладающий еще большими возможностями.

````
plt.scatter(x, y, marker='o')
````

Продемонстрируем более сложные возможности. Чтобы видеть перекрывающиеся результаты, можно использовать параметр непрозрачности `alpha`:

````
rng = np.random.RandomState(0)
x = rng.randn(100) # 100 случайных точек нормально распределенных
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar() # отображаем цветовую шкалу
````

Рассмотрим набор данных Iris из библиотеки Scikit-Learn (данные о длине и ширине лепестков и чашелистиков)

````
from sklearn.datasets import load_iris

iris = load_iris()

features = iris.data.T
plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
````

На этой диаграмме рассеяния координаты точек соответствуют длине и ширине чашелистика, размер точки - ширине лепестков, а цвет - конкретной разновидности цветка (их 3)

### Heatmaps

Нeatmaps - двумерные диаграммы, в которых цветом закодирована величина, зависящая от 2-х измерений.

Можно опционально выводить аннотации, например численные значения величин.

Пример из документации показывает, как это сделать.

````
# сделаем массив демо-данных
vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])
````

Простейший вариант:

````
plt.imshow(harvest)
plt.colorbar();
````

Более сложный вариант - с аннотациями и метками

````
fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(harvest)

# We want to show all ticks...
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))
# ... and label them with the respective list entries
ax.set_xticklabels(farmers)
ax.set_yticklabels(vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j], ha="center", va="center", color="w")

ax.set_title("Harvest of local farmers (in tons/year)")
fig.tight_layout()
````

Подробности см. в [документации](https://matplotlib.org/stable/gallery/images_contours_and_fields/image_annotated_heatmap.html)

### Субграфики

````
plt.figure(figsize=(12, 6)) # задаем размеры всего рисунка
for i in range(1, 7): # в цикле создаем 6 субграфиков.
    plt.subplot(2, 3, i)  # 2 строки, 3 колонки субграфиков. i - номер субграфика
    plt.text(0.5, 0.5, "(2, 3, %d)" % i, fontsize=18, ha='center') # вывод пойдет в текущий субграфик
````

Существует альтернативный способ создания субграфиков с помощью функции `plt.subplots()`, который в некоторых случаях может быть удобнее.

````
fig, ax = plt.subplots(2, 3, figsize=(12, 6), sharex='col', sharey='row') # ось x - общая для всех колонок, ось y - для всех строк.
ax.shape
````

````
for i in range(2):  # индексация с 0!
    for j in range(3):
        ax[i, j].text(0.5, 0.5, str((i, j)), fontsize=18, ha='center')
fig # выводим весь рисунок
````

## Seaborn

К недостаткам библиотеки Matplotlib можно отнести то, что она не ориентирована на работу с объектами DataFrame библиотеки Pandas, так как последняя вышла на 10 лет позже Matplotlib. Библиотека [Seaborn](http://seaborn.pydata.org/) - решение этой проблемы. Она представляет собой высокоуровневый интерфейс ([API](http://seaborn.pydata.org/api.html)) к Matplotlib.

````
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline
import numpy as np
import pandas as pd
````

Создадим тестовые данные случайных блужданий (Берем сумму случайных чисел нарастающим итогом)

````
# Create some data
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
````

Выведем простой график:

````
# Plot the data with Matplotlib defaults
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
````

Посмотрим как можно те же данные вывести в seaborn. Сначала импортируем библиотеку.

````
import seaborn as sns
sns.set_style("darkgrid")
````

Вывести помощь по установке стилей можно так:

````
sns.set_style?
````

Повторим тот же самый код:

````
# same plotting code as above!
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
````

````
x = np.linspace(-5, 5, 100)
for style in ['darkgrid', 'whitegrid', 'dark', 'white']: # перебираем стили графиков
    sns.set_style(style) 
    plt.plot(np.cos(x))
    plt.title(f"Style: {style}")
    plt.show()
sns.set_theme(style="darkgrid") # восстановим стиль
````

### Гистограммы, KDE, плотности.

Гистограммы:

Создадим массив демо-данных из 2000 двумерных случайных данных (x, y) со средним значением (0, 0) и матрицей ковариаций 
```
[[5, 2], 
 [2, 2]]
```

````
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data
````

Преобразуем массив в pandas DataFrame - seaborn ориентирован на данные pandas:

````
data = pd.DataFrame(data, columns=['x', 'y'])
data
````

Построим гистограмму данных средствами matplotlib:

````
for col in ('x', 'y'):
    plt.hist(data[col], alpha=0.5)
````

Вместо гистограммы можно получить гладкую оценку распределения путем ядерной оценки плотности распределения KDE (kernel density estimation) c помощью функции Seaborn `sns.kdeplot`:

````
for col in ('x', 'y'):
    sns.kdeplot(data[col], shade=True)
````

Можно получить двумерное распределение если передать `kdeplot` сразу двумерный набор:

````
sns.kdeplot(x="x", y="y", data=data, shade=True);
````

еще один способ вывести двумерное распределение:

````
sns.displot(x="x", y="y", data=data, kind="kde");
````

Посмотреть совместное и частные распределения можно с помощью функции jointplot:

````
with sns.axes_style('darkgrid'):
    sns.jointplot(x="x", y="y", data=data, shade=True, kind='kde');
````

Библиотека Seaborn имеет собственный набор демо-датасетов. Загрузим один из них:

````
penguins = sns.load_dataset("penguins")
penguins
````

Построим диаграммы распределения длин и ширин клюва у разных видов пингвинов:

````
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", height=8);
````

### Heatmaps

"тепловые карты" - диаграммы, изображающие двумерные распределения, цвет обозначает величину, зависящую от 2-х переменных, например матрицу корреляций.

Получить подсказку:

````
sns.heatmap?
````

Построим матрицу взаимных корреляций величин из 4-х столбцов:

````
corr = penguins.corr()
corr.head()
````

Создадим и нарисуем палитру, которую будем использовать для построения heatmap

````
cmap = sns.diverging_palette(0, 250, sep=1, n=20)
sns.palplot(cmap) # diverging_palette - расходящаяся палитра между двумя цветами
                                                        # 0 - нижняя граница цвета, 250 - верхняя граница цвета
                                                        # sep - размер промежуточного региона, n - количество цветов в палитре
````

Получение справок:

````
sns.diverging_palette?
````

````
sns.palplot?
````

Выведем матрицу корреляций в виде "тепловой карты". Величины близкие к 1 или -1 означают, что 2 данных показателя взаимозависимы. Величины близкие к 0 показывают что показатели не коррелируют (взаимно независимы)

````
plt.figure(figsize=(7,6))
sns.heatmap(corr, cmap=sns.color_palette(cmap), vmin=-1, center=0, annot=True) 
````
