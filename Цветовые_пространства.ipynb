```python
# import the necessary packages
# импортируем необходимые модули
import cv2
from google.colab.patches import cv2_imshow
```

```python
# construct the argument parser and parse the arguments
# создадим парсер и укажем аргументы
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--image", required=True, help="Path to the image")
# args = vars(ap.parse_args())
# пока мы используем Jupyter Notebooks мы можем заменить наш парсер
# явными аргументами и значениями
args = {
	"image": "/content/Lesson_17.Color_spaces/zebra.png",

}
```

```python
# load the original image and show it (RGB)
# загрузим исходное изображение и покажем его (RGB)
image = cv2.imread(args["image"])
cv2_imshow(image)
print("RGB")
```

```python
# loop over each of the individual channels and display them
# переберём каждый из отдельных каналов и отобразим их
for (name, chan) in zip(("B", "G", "R"), cv2.split(image)):
	print(name)
	cv2_imshow(chan)
 
```

```python
# convert the image to the HSV color space and show it
# преобразуем изображение в цветовое пространство HSV 
# и выведем его на экран
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
print("HSV")
```

```python
# loop over each of the invidiaul channels and display them
# переберём каждый из отдельных каналов и отобразим их
for (name, chan) in zip(("H", "S", "V"), cv2.split(hsv)):
	print(name)
	cv2_imshow(chan)
```

```python
# convert the image to the L*a*b* color space and show it
# преобразуем изображение в цветовое пространство L*a*b* 
# и выведем его на экран
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
cv2_imshow(lab)
print("L*a*b*")
```

```python
# loop over each of the invidiaul channels and display them
# переберём каждый из отдельных каналов и отобразим их
for (name, chan) in zip(("L*", "a*", "b*"), cv2.split(lab)):
	print(name)
	cv2_imshow(chan)
```

```python
# show the original and grayscale versions of the image
# покажем исходное изображение и
# в градациях серого
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(image)
print("Original")
```

```python
cv2_imshow(gray)
print("Grayscale")
```
