```python
# import the necessary packages
# импортируем необходимые модули
import numpy as np
import imutils
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# construct the argument parser and parse the arguments
# создадим парсер и укажем аргументы
# ap = argparse.ArgumentParser()
# ap.add_argument("-i", "--image", required=True, help="Path to the image")
# args = vars(ap.parse_args())
# пока мы используем Jupyter Notebooks мы можем заменить наш парсер
# явными аргументами и значениями
args = {
	"image": "/content/Lesson_05.Rotation/tigers.png",

}
```

```python
# load the image and show it
# загрузим изображение и покажем его
image = cv2.imread(args["image"])
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.title("Original")
```

```python
# grab the dimensions of the image and calculate the center of the image
# возьмём размеры изображения и вычислим его центр
(h, w) = image.shape[:2]
(cX, cY) = (w / 2, h / 2)

# rotate our image by 45 degrees
# повернём изображение на 45 градусов
M = cv2.getRotationMatrix2D((cX, cY), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(rotated)
plt.title("Rotated by 45 Degrees")
```

```python
# rotate our image by -90 degrees
# повернём изображение на -90 градусов
M = cv2.getRotationMatrix2D((cX, cY), -90, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(rotated)
plt.title("Rotated by -90 Degrees")
```

```python
# rotate our image around an arbitrary point rather than the center
# повернём изображение вокруг произвольной точки, а не вокруг центра
M = cv2.getRotationMatrix2D((cX - 50, cY - 50), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
plt.imshow(rotated)
plt.title("Rotated by Offset & 45 Degrees")
```

```python
# finally, let's use our helper function in imutils to rotate the image by
# 180 degrees (flipping it upside down)
# наконец, давайте воспользуемся нашей вспомогательной функцией
# в imutils, чтобы повернуть изображение на 180 градусов
# (перевернуть его вверх ногами)
rotated = imutils.rotate(image, 180)
plt.imshow(rotated)
plt.title("Rotated by 180 Degrees")
```
